# Figuring out how much flux after an introduction of excess D.

```python
import matplotlib.pyplot as plt
import scipy.interpolate as spi
import scipy.integrate as spint
import numpy as np
```

```python
times = np.array([0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0, 10000.0, 100000.0, 1.0e6])
times_log = np.logspace(-4, 5, 10)
print(times_log)
ex_flux = np.array([2.2821324829610603, 7.521132613374448, 363.3682993379002, 9180.325121268588, 13883.23209124, 10289.252694209437, 
                    6635.110452073766, 2367.0916738786223, 90.87325212969836, -1.6385826840901245])
```

```python
plt.figure(figsize=(9,6))
plt.loglog(times, ex_flux, marker=".")
plt.xlabel("Time (s)")
plt.ylabel(r"Flux (cm$^{-2}$s$^{-1}$)")
plt.show()
```

```python
# Hand attempt at integration, centering the intervals on the points.
# dt = np.asarray([((5*t) - (5*(t-1))) for t in times_log])
dt = 5*times_log[1:] - (5*times_log[:-1])
result = sum(ex_flux[:-1] * dt)
print("{:1e}".format(result))
```

```python
result2 = spint.trapz(ex_flux, x=times)
print("{:.8e}".format(result2)) # this should be in #/cm^2
```
